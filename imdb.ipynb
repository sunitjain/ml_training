### Let us start with some Basic Analysis 

```python
df["genres"].value_counts()
```

```python
seperate_genre='Action','Adventure','Animation','Biography','Comedy','Crime','Drama','Fantasy','Family','History','Horror','Music','Musical','Mystery','Romance','Sci-Fi','Sport','Thriller','War','Western'
for genres in seperate_genre:
    df2 = df['genres'].str.contains(genres).fillna(False)
    print 'The total number of movies with ',genres,'=',len(df[df2])
    f, ax = plt.subplots(figsize=(15, 6))
    sn.countplot(x='title_year', data=df[df2], palette="BuGn_d");
    plt.title(genres)
    compare_movies_rating = ['duration', 'num_user_for_reviews','gross']
    for compare in compare_movies_rating:
        sn.jointplot(x='imdb_score', y=compare, data=df[df2], alpha=0.8, color='b', size=8)
        plt.title(genres)
```
