```python
#Import the necessary libraries

import pandas as pd
import numpy as np
from sklearn.cross_validation import train_test_split
```

```python
#Load the dataset

white = pd.read_csv('winequality-white.csv')
white = white['fixed acidity;"volatile acidity";"citric acid";"residual sugar";"chlorides";"free sulfur dioxide";"total sulfur dioxide";"density";"pH";"sulphates";"alcohol";"quality"'].str.split(';', 11, expand=True)
white.columns = ['fixed acidity','volatile acidity','citric acid','residual sugar','chlorides',"free sulfur dioxide","total sulfur dioxide","density","pH","sulphates",'alcohol','quality']
white = white.apply(pd.to_numeric, errors='ignore')
```

```python
#Split the data into train, test and then train features and quality

train_data ,test_data = train_test_split(white,test_size=0.2)
train_features = train_data.iloc[:, :-1].copy()
train_quality = train_data.iloc[:, -1:].copy()
```

```python
#Define intercept, merge with train feature df

intercept = np.ones((train_features.shape[0], 1))
train_features = np.concatenate((intercept, train_features), axis=1)
```

```python
m = len(train_features)
```

```python
#Split the data into test features and quality

test_features = test_data.iloc[:, :-1].copy()
test_quality = test_data.iloc[:, -1:].copy()
```

```python
#Define intercept, merge with test feature df

intercept = np.ones((test_features.shape[0], 1))
test_features = np.concatenate((intercept, test_features), axis=1)

test_features = (test_features - np.mean(test_features))/np.std(test_features)
train_features = (train_features - np.mean(train_features))/np.std(train_features)

theta = np.zeros((train_features.shape[1]))
grad = np.zeros((train_features.shape[1]))
theta = theta.reshape(train_features.shape[1],1)

#whitewine["balfac"]=np.full((len(whitewine),1),1)
#msk = np.random.rand(len(whitewine)) < 0.7
#train_data = whitewine[msk]
#test_data = whitewine[~msk]
#train_data_x = train_data.loc[:, whitewine.columns != 'quality']
#train_data_y = train_data.loc[:,whitewine.columns == 'quality']
#train_data_x = train_data_x.values
#train_data_y = train_data_y.values
```

```python
#Define Gradient Descent

def gradient(w, data_x, data_y):
    y_estimated = ((data_x).dot(w))
    grad = (1.0/len(data_x))*((data_x).T.dot(y_estimated-data_y))
    cost = (1.0/(2*len(data_x)))*(np.sum(np.square(y_estimated-data_y)))
    return grad, cost
```

```python
alpha = 0.0001
#tolerance = 0.1
iters = 1
cost = 1000
```

```python
#Run iterations to reduce cost function

while True:
    grad, cost_new = gradient(theta, train_features, train_quality)
    theta_new = theta - (alpha*grad)
    #Stop Condition
    #if abs(cost - cost_new) < tolerance:
    if iters > 25000:
        print "Converged."
        break
    #if iters%100 == 0:
    #     print iters, cost
    iters = iters +1
    theta = theta_new
    cost = cost_new
#print iters
```

```python
#Predict test quality

def predict(w, x, y):
    h = x.dot(w)
    error = np.sum(abs(h-y))
    return h, error
y, error = predict(theta, test_features, test_quality)
print("Predicted values of quality:", np.around(y))
```
